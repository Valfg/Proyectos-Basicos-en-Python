# Convertidor de Temperaturas
## Este script es un programa que convierte temperaturas entre tres unidades: Celsius, Fahrenheit y Kelvin

```python
"""Toma una temperatura en grados Celsius y la convierte a Fahrenheit y Kelvin.
Devuelve un diccionario con las temperaturas en las tres unidades."""

def Celsius_(Celsius):
    return  {
        "celsius" : Celsius,
        "farenheit": (Celsius * 9/5)+32,
        "kelvin": Celsius + 273.15}

"""Toma una temperatura en Fahrenheit, la convierte a Celsius y luego a Kelvin.
Devuelve un diccionario con las temperaturas en las tres unidades."""

def Farenheit_(Farenheit):
    Celsius= (Farenheit -32) * 5 / 9
    return { 
        "celsius":Celsius,
        "farenheit": Farenheit,
        "kelvin": Celsius + 273.15}

"""Toma una temperatura en Kelvin, la convierte a Celsius y luego a Fahrenheit.
Devuelve un diccionario con las temperaturas en las tres unidades."""

def Kelvin_(Kelvin):
    Celsius= Kelvin -273.15
    return { 
        "celsius": Celsius,
        "farenheit": (Celsius * 9/5)+32,
        "kelvin": Kelvin}

""" FUNCIÓN PRINCIPAL: 
Recibe como entrada la unidad de la temperatura y la temperatura en sí.
Usa un diccionario para seleccionar la función correcta según la unidad ingresada.
Llama a la función correspondiente y devuelve el resultado. """

def convertidor (unidad_temperatura, temperatura):
    convertidores= { 
        1: Celsius_,
        2: Farenheit_,
        3: Kelvin_}
   
    if unidad_temperatura in convertidores:
        return convertidores[unidad_temperatura] (temperatura) 


# Entrada del usuario: Pide al usuario que introduzca la temperatura y su unidad (1 para Celsius, 2 para Fahrenheit, 3 para Kelvin).

temperatura=float(input("Ingrese la temperatura que desea convertir: "))
unidad_temperatura= int(input("Ingrese el número correspondiente a la unidad de la temperatura que ingresó : 1 (Celsius), 2 (Farenheit), 3 (Kelvin)"))

# Si la unidad no es válida, vuelve a pedirla utilizando un bucle WHILE.
while unidad_temperatura not in range(1,4):
    unidad_temperatura= int(input("El número no es correcto!!! Por favor ingrese: 1 (Celsius), 2 (Farenheit), 3 (Kelvin)"))

""" Conversión y resultado: Llama a la función convertidor para obtener las conversiones.
Imprime un diccionario con los valores en Celsius, Fahrenheit y Kelvin."""

resultado= convertidor(unidad_temperatura, temperatura)
print(resultado)
```

# Contador de palabras de un texto

## Este script cuenta la frecuencia de cada palabra de un texto ingresado por el usuario.

```python
import re # Permite trabajar con expresiones regulares, se utiliza para limpiar el texto de caracteres especiales.

# Se solicita al usuario que introduzca un texto. Convierte el texto a minúsculas con .lower() para evitar problemas de distinción entre mayúsculas y minúsculas.
texto=str(input("Ingrese un texto:")).lower() 

def contador(texto):
    texto_nuevo= re.sub(r'[^\w\s]', '', texto)  # Se utiliza el patrón r'[^\w\s] para reemplazar cualquier carácter que no sea una letra, número o espacio .
   
    palabra=texto_nuevo.split() # Separa el texto en una lista de palabras basada en los espacios.
    
    dic_frecuencia={} # Crea un diccionario dic_frecuencia para almacenar cada palabra como clave y su frecuencia como valor.
    
    # Utiliza un bucle FOR para recorrer cada palabra en la lista.
    for palabra in palabra:
        if palabra in dic_frecuencia:
            dic_frecuencia[palabra]+=1 # Si la palabra ya está en el diccionario, incrementa su contador.
        else:
            dic_frecuencia[palabra] =1 # Si no está, agrega la palabra al diccionario con un contador inicial de 1.
    
    # Recorre el diccionario y para cada palabra, imprime su frecuencia.
    for palabra in dic_frecuencia:
        frecuencia= dic_frecuencia[palabra]
        print(f"La palabra '{palabra}' tiene una frecuencia de {frecuencia}")

contador(texto)
        
```

# Diccionario inverso

## Este script invierte un diccionario convirtiendo claves en valores y viceversa.
## Verifica si hay valores duplicados en el diccionario original; si existen, no realiza la inversión y devuelve un mensaje de error, asegurando que no se pierda información.

```python
# La función toma un diccionario como entrada
def dic_invertido(diccionario):
    resultado= {} # Almacena la versión invertida del diccionario
    for k,v in diccionario.items(): # Usa un bucle for para iterar por las claves (k) y valores (v) del diccionario.
        if v in resultado: 
            print("Error: múltiples llaves para un mismo valor")
            return None
        """Si el valor v ya está en el diccionario "resultado" (es decir, dos claves diferentes tienen el mismo valor), 
        mprime un mensaje de error y termina la función con return None."""
        resultado [v] = k # Si no hay conflictos, agrega "v" como clave y "k" como valor al diccionario "resultado"
    return resultado  # Devuelve el diccionario con las claves y valores invertidos.        

dic_inv = {"Zapatos": 1, "Pantalón": 2, "Sombrero": 3}
dic_inv_1={"Zapatos": "azules", "Pantalón": "verde", "Sombrero": "verde"}
diccionario_invertido= dic_invertido(dic_inv)
if diccionario_invertido is not None:
    print(diccionario_invertido)
diccionario_invertido_1= dic_invertido(dic_inv_1)
if diccionario_invertido_1 is not None:
    print(diccionario_invertido_1)
```

# Diccionario inverso con duplicados

- El objetivo del script es invertir un diccionario, intercambiando claves y valores. 
- Si varias claves comparten el mismo valor, agrupa esas claves en una lista. 
- Es útil para analizar relaciones inversas y manejar datos con valores no únicos.

```python
# Toma un diccionario como entrada y crea uno nuevo (dic_inv_2) donde los valores originales se convierten en claves, y las claves originales en valores.
def dic_inv_plus (diccionario):
    dic_inv_2= {}
    for k, v  in diccionario.items():
        if v in  dic_inv_2: # Si el valor v ya existe como clave en el nuevo diccionario, se debe manejar el caso especial de múltiples claves con el mismo valor.
            if isinstance( dic_inv_2 [v], list):
                [dic_inv_2[v].append(k)] # Si ya hay una lista asociada al valor, se añade la nueva clave.
            else:
                dic_inv_2[v] = [dic_inv_2[v], k] # Si no es una lista, se convierte en una lista para incluir ambas claves.
        else: # Si el valor no existe aún como clave en el nuevo diccionario
            dic_inv_2[v] = k  # Se agrega el valor original como nueva clave, con la clave original como su valor.
    return  dic_inv_2

dic_inv = {"Zapatos": 1, "Pantalón": 2, "Sombrero": 3}
inv_dic=dic_inv_plus(dic_inv)
print(inv_dic)
dic_inv_1={"Zapatos": "azules", "Pantalón": "verde", "Sombrero": "verde"}
inv_dic_plus=dic_inv_plus(dic_inv_1)
print(inv_dic_plus)
```

# Conversión de tipos de datos

### Este script toma una lista de datos con elementos variados (cadenas, números, listas y tuplas) y realiza dos operaciones principales:
- Desglosar los datos para crear una lista plana, eliminando las anidaciones de listas o tuplas.
- Separar los elementos en dos listas: una con elementos que se pueden convertir a números flotantes (float) y otra con los que no.

```python
datos=['1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

# Extrae los datos por separado y los agrega a una lista.
def extraer(listades):
    lista_desglosados= []
    for x in listades:
        if isinstance(x, (list,tuple)):    
            lista_desglosados.extend(x) # Si el elemento es una lista o tupla: Usa extend para añadir sus elementos a la lista lista_desglosados.
        else:
            lista_desglosados.append(x) 
    return lista_desglosados
lista_correcta= extraer(datos)

# Divide los elementos en dos listas, una con los datos que se pueden convertir a float y otra con los que no.
def convertir(f):
    lista_float=  []
    lista_no_float=  []
    for x in f :
        try:
            float_item=float(x) # Intenta convertir un elemento a un número flotante.
            lista_float.append(float_item) # Si puede convertir el elemento a un número flotante lo agrega a lista_float.
        except ValueError:
            lista_no_float.append(x) # Si no puede convertir el elemento a número flotante lo agrega a la lista_no_float.
    return lista_float, lista_no_float
convertir(lista_correcta)
```

# Contador y ordenador de palabras de un texto.
### Este script procesa un archivo de texto y realiza las siguientes operaciones:
- Cuenta el número total de palabras en el texto.
- Crea un diccionario con la frecuencia de cada palabra.
- Agrupa las palabras según su letra inicial.
- Convierte el resultado en formato JSON para una visualización más organizada.

```python

import re # Permite trabajar con expresiones regulares, se utiliza para limpiar el texto de caracteres especiales.
import json # Permite convertir el resultado en un formato legible como JSON.
from collections import defaultdict # Una variante del diccionario que permite crear diccionarios anidados fácilmente.

def contador_dic():
    nombre_del_archivo= input("Introduzca el nombre del archivo (con extensión): ")
    with open (nombre_del_archivo, 'r', encoding='utf-8') as archivo: # Abre el archivo en modo de lectura ('r') con codificación utf-8 para evitar problemas con caracteres especiales.
        contenido=archivo.read() # Lee todo el contenido del archivo y lo devuelve como una cadena.
    return contenido 

def contador():
    contenido= contador_dic()
    contenido_palabras= re.sub(r'[^\w\s]', '', contenido).lower() # Elimina todo lo que no sea texto alfanumérico o espacios ([^\w\s]).
    palabras=contenido_palabras.split() # Usa split() para dividir el texto en una lista de palabras basándose en los espacios.
    dic_frecuencia={} 
    for palabra in palabras:
        if palabra in dic_frecuencia:
            dic_frecuencia[palabra]+=1  # Incrementa la frecuencia si ya existe.
        else:
            dic_frecuencia[palabra] =1 # Añade la palabra con frecuencia 1.
    palabras_agrupadas_inicial= defaultdict(dict)
    for palabra, frecuencia in sorted(dic_frecuencia.items()):
        letra_inicial=palabra[0] # Obtiene la primera letra utilizando la primera posición del índice.
        palabras_agrupadas_inicial[letra_inicial][palabra] =frecuencia
    
    palabras_agrupadas=dict(palabras_agrupadas_inicial)
    print (F"El texto tiene {len(palabras)} palabras.") # Devuelve la cantidad de palabras que tiene el texto.
    return palabras_agrupadas

resultado= contador()
salida= json.dumps(resultado, indent=2, ensure_ascii=False)
print(salida) # Devuelve el diccionario con las palabras y su frecuencias, ordenadas alfabéticamente.

```
